## Frequency Distributions

Very often used to create basic Features on Texts. In their most basic form, they create a 1-'Bag of Words' count, that just displays indivual word frequencies. 

In large Corpora it is likely that we want to focus on the words with the highest discriminatory capabilities. This is done by droping 'stopwords' contained in special Dictionaries for each language from the Text.

```python
from nltk import FreqDist
text1 = 'This s a simple text that i stored here to calculate the first frequency distribution on it'.split(' ')
FreqDist(text1)
```

```python
vocabulary = FreqDist(text1).keys()
print(vocabulary)
```

```python
FreqDist(text1).plot()
```

```python
text2 = "Well what can i say, i just really like to repeat myself. I like to hear myself repeating what i have just repeated, because i like to repeat myself.".split(' ')
freq_2 = FreqDist(text2)
vocab_2 = FreqDist(text2).keys()
print(freq_2)
freq_2.plot()
```

```python
freq_2.plot(20,cumulative=True)
```

```python
# Total number of samples
print(freq_2.N())

# Get a frequency count for a specific word
print(freq_2['like'])

# Or the percentage
print(freq_2.freq('like'))

# Sample with max frequency
print(freq_2.max())
```

```python
# Tabulate the result of the Distribution
freq_2.tabulate()
```

```python
# Plot the result
plt.title('Frequency Based Line Plot')
freq_2.plot()
plt.title('Cumulative Plot')
# Plot it as cumulative in desc order
freq_2.plot(cumulative=True)
```

We can use a large sample of built in python String Commands

```python
s = 'I am a regular string, that contains both grammatical information as well as punctuation! Really... Don\'t you trust me??'
print(s)
```

```python
# Check for Beginning
print( s.startswith('I') )

# Check for Ending
print( s.endswith('??') )

# Check for containment
print( 'grammatical' in s )

# Check for Casing
print( s[0].islower() )
print( s[0].isupper() ) 
print( s[0].istitle() )

# Check for Type
print( s[0].isalpha() )
print( s[0].isalnum() )
print( s[0].isdigit() )
```

### Descriptive Statistics used on Text

```python
# count based Features

# Length of the overall text
print( len(text2) )

# Number of unique tokens(individual string elements)
print(len(set(word.lower() for word in text2)))

# Number of unique word-tokens
print( len(set(word.lower() for word in text2 if word.isalpha())) )
```
