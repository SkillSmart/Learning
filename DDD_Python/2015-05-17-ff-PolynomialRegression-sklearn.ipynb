# Implementing a polynomial regression using Sklearn 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Generate random 2polynomial function
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + 2 + np.random.randn(m,1)
```

```python
plt.scatter(X,y)
```

```python
def plot_prediction(model, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    model_trn = model()
    model_trn.fit(X_train,y_train)
    predictions = model_trn.predict(X_test)
    # Draw the plot
    plt.plot(X_train, y_train, 'bo', label='train')
    plt.plot(X_test, predictions, '--rx', label='validatio')
```

```python
def plot_learning_curve(model, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_test_predict = model.predict(X_test)
        train_errors.append(mean_squared_error(y_train_predict, y_train[:m]))
        val_errors.append(mean_squared_error(y_test_predict, y_test))
    plt.plot(np.sqrt(train_errors), 'r-+', linewidth=2, label='train')
    plt.plot(np.sqrt(val_errors), 'b-', linewidth=3, label='val')
    plt.xlabel('Training Set Size')
    plt.ylabel('RMSE')
    plt.legend()
    plt.grid()
```

Polynomial Fitting using 'PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
```

```python
X_poly = poly_features.fit_transform(X)
```

```python
print(X[0])
print(X_poly[0])
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
# Display the prediction
plot_prediction(LinearRegression, X_poly, y)
```

```python
# Display training s
lin_reg = LinearRegression()
plot_learning_curve(lin_reg, X,y)
```

The Models capability of fitting the data levels around 20 Samples, after which adding new data does not increase the model fit any further.<br>
We have to add degrees of freedom to the model to enable it to learn more complex features, but adding too many will result in a severe case of overfitting.

```python
from sklearn.pipeline import Pipeline

polynomial_regression = Pipeline((
    ('poly_features', PolynomialFeatures(degree=10, include_bias=False)),
    ('sgd_reg', LinearRegression()),
))

plot_learning_curve(polynomial_regression, X, y)
```

With a 10-degreee polynomial function, the model is able to learn all features in the dataset around 48 Items. After this the model progresses slowly keeping the validation error always larger than the training error. This is a sign of overfitting

## Methods of Regularization for polynomial linear Models

First, the Ridge Regression Model adds a penalizing factor to the linear Regression Model to force small weights on the solution.

```python
#Ridge Regression penalizes a model for large weights, and learns to find the model with the best fit at the min assumption
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1, solver='cholesky')
ridge_reg.fit(X,y)
ridge_reg.predict([[1.5]])
```

```python
# Implementing Ridge using SGD
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(penalty='l2')
sgd_reg.fit(X,y.ravel())
sgd_reg.predict([[1.5]])
```

### Lasso Regression (Least Absolute Shrinkage and Selection Operator Regression)

Tends to set the least factor weights to zero, thereby effectively reducing the dimension of the solution.<br>
Thereby automatically performing a feature selection and outputting a sparse representation

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X,y)
lasso_reg.predict([[1.5]])
```

### Elastic Net

Builds the middle ground between Ridge and Lasso, adding both terms in a specified ratio.

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X,y)
elastic_net.predict([[1.5]])
```

### 'Early Stopping': find the best model trough iterive BGD/SGD

Take the settings for the model that had the best Validation Set performance, and stop trainig there. 
Easy to use!

```python
from sklearn.base import clone
sgd_reg = SGDRegressor(n_iter=1, warm_start=True, penalty=None,
                      learning_rate='constant', eta0=0.0005)

minimum_val_error = float('inf')
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X, y.ravel())
    y_val_predict = sgd_reg.predict(X)
    val_error = mean_squared_error(y_val_predict, y)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
```

```python
best_model.fit(X,y.ravel())
```

## Using Linear Functions as Classifiers

Linear Models can computes a weighted sum of all its inputs, but instead of outputing a result directly, it learns to maximize values between 0 and 1 to differentiate between 2 classes of outputs. 
This holds true beyond a binary classification into n-dimensional classification examples. Often used as the output layer of a neural network classifier.

```python
# Building a classifier for the Iris Dataset
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target)
```

```python
print(X_train.ndim)
print(y_train.ndim)
```

```python
from sklearn.linear_model import LogisticRegressionCV

logr_cv = LogisticRegressionCV()
logr_cv.fit(X_train, y_train)
```

Looking at only one dimension of the dataset, we can visualize the decission boundary the model fits to select between Class or not class for a specific Flower Type.

```python
from sklearn.linear_model import LogisticRegression

# Extract and train the model only on the petal width feature
X = iris['data'][:,3:]
y = (iris['target']==2).astype(np.int)

# Fitting the Model
log_reg = LogisticRegression()
log_reg.fit(X,y)

# Let's visualize the decission boundary the model fitted to the data
X_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], 'b--', label='Not Iris-Virginica')
plt.xlabel('Petal width (cm)')
plt.ylabel('Propability')
plt.legend(loc='center left')
```

The Model learned a logistic function over the Petal width Dimension.

To support multi class classification directly, instead of training multiple Linear Models, we can turn to using the 'Softmax Regression'

### Multinomial Logistic Regression

The model learns a linear combination of input features that give the highest propability for a given class label y.
Learning happens trough reduction of the 'Cross Entropy' Cost function, with an interative Optimizer like SGD, etc..

```python
X = iris['data'][:, (2,3)]
y = iris['target']

softmax_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', C=10)
softmax_reg.fit(X,y)
```

```python
# Testing the model
softmax_reg.predict([[5,2]])
```

```python
softmax_reg.predict_proba([[5,2]])
# Showing the respective propabilty scores for each class
```
